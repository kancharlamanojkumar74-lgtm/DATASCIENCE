Write a function that accepts a list of numbers and returns the largest number.

```python
def find_largest(numbers_list):
  return max(numbers_list)

my_list = [1,2,3,4,5]
print(find_largest(my_list))
```

Write a function that takes a sentence and returns the number of vowels in it.

```python
def Vowels(sentence):
  Vowels="aeiou"
  for ch in sentence:
    if ch.lower() in Vowels:
      print(ch)
sentence="hi this is manoj"
Vowels(sentence)
```

Write a function that accepts student marks and returns grade (A/B/C/Fail).

```python
def Grade():
  marks=int(input("enter marks"))
  if marks>=90:
    return"grade A"
  elif marks>=80:
      return"grade B"
  elif marks>=70:
        return"grade c"
  else:
    return "fail"
print(Grade())
```

Write a function that takes a number and returns whether it is prime or not.

```python
def prime_or_Not():
  number=int(input("enter a number"))
  if number>1:
   for i in range(2,number):
      if number % i==0:
        return "not prime"
      return "prime"
  else:
    return "not prime"
print(prime_or_Not())
```

Lambda Function  
1. Use lambda to sort a list of tuples based on the second value.


```python
data = [("apple", 3), ("banana", 1), ("orange", 2)]
sorted_list = sorted(data, key=lambda x: x[1])
print(sorted_list)
```

 Use lambda to find the last digit of a number.



```python
last_digit=lambda num:num%10
print(last_digit(1289))
```

Map Function
1. Use map() to convert the list of names into uppercase.

```python
def convert_upper(name):
  return name.upper()
names=["manoj","kumar","anil","ram"]
upper_names= __builtins__.list(__builtins__.map(convert_upper,names))
print(upper_names)
```

. Use map() to calculate the square root of numbers in a list.

```python
import math

def square_root(x):
    return math.sqrt(x)

numbers = [2, 4, 5, 6]

square_roots = __builtins__.list(__builtins__.map(square_root, numbers))

print(square_roots)
```

Given prices list, use map() to add 18% GST to each value.

```python
prices=[10000,20000,3000]
gst=list(map(lambda x:x+x*18/100,prices))
print(gst)
```

Use map() to extract the first character from each word in the list.

```python
List=["apple","Ball","cat"]
extract=list(map(lambda x:x[0],List))
print(extract)
```

 Convert list of temperatures from Fahrenheit to Celsius using map()

```python
temp=[70,73,90]
celsius=list(map(lambda x:(x-32)*5/9,temp))
print(celsius)
```
